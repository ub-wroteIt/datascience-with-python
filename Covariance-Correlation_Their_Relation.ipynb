# Covariance, Correlation and Their relation

covarience measures how much two varibles very together.its is similar to varience where varience tells you about single variable varience coverience tells you about how two varibles very in tandem.

Lets say we have an Object detection Application used for face detection that had go to web service and findout the images of celebraties and other details of application, We are interested in finding the correlation between page speeds and how much time user will spend to mark person's face and identify it.

The idea behind this is treat each varible as vector of deviation from mean, compute the dot product of both. We can think of this as angle between the vectors which simulates the measure of similarity between the two varibles.

## Lets gear up our environments

```python
%matplotlib inline

import numpy as np
from pylab import *

def get_deviation_mean(elements):
    emean = mean(elements)
    return [elementsi - emean for elementsi in elements]
def get_covarience(x,y):
    n = len(y)
    return dot(get_deviation_mean(x),get_deviation_mean(y))/ (n-1)

pageSpeed = np.random.normal(3000,1000,200)
```

### When it has correlation

```python
detectionTime = np.random.normal(10000,1500,200)*pageSpeed

scatter(pageSpeed, detectionTime)
get_covarience(pageSpeed, detectionTime)
```

### When it has No Real Correlation

To simulate this this lets make page speed and detection time totally random and independent of each other which will not have any corelation.

```python
detectionTime = np.random.normal(10000,1500,200)

scatter(pageSpeed, detectionTime)
get_covarience(pageSpeed, detectionTime)
```

### But How to interprete this?

Coverience is sensitive to units used which makes it difficult to interpret.

And that where correlation helps it normalizes everthing by their standard deviation , giving values from -1 (Negative correlation) to 1 (Positive correlation ) which are easier to understand.

```python
def get_correlation(x, y):
    stddevx = x.std()
    stddevy = y.std()
    return get_covarience(x,y) / (stddevx * stddevy)

get_correlation(pageSpeed, detectionTime)
```

```python
detectionTime = np.random.normal(10000,1500,200)*pageSpeed

get_correlation(pageSpeed, detectionTime)
```

### With NumPy

NumPy can do all this smoothly.It runs the matrix of correlation coefficients.

```python
np.corrcoef(pageSpeed, detectionTime)
```

we cant force perfect correlation by overfitting a totally linear relationship.But its close enough to tell us that there exist good correlation.

```python
detectionTime =  pageSpeed*3
scatter(pageSpeed, detectionTime)

get_correlation (pageSpeed, detectionTime)
```
