# Decision Trees

### 1. It construct a flowchart / Tree to help us decide a classfication for something with machine learning.  

### 2. Another form of supervised learning.

### Its Application

1. Hiring system which filter out resumes based on historical hiring data. Train a decision tree with historical hiring data and arrive at a system for predicting whether a candidate will get hired based on it.
2. Classifying cutomer for loan approval.
3. Face recognition to identify a pattern on the face which recognizes a criminal.

### How it Work

Fancy term : ID3
+ It uses Entropy and Information Gain to construct a decision tree.
+ It is Greedy Algorithm
    * as it goes down the tree, it just picks the decision that reduce entropy the most at that stage.
    * It might not actually result in an optimal tree.

### Entropy

+ It measures the data set's disorder or InPurity - how same or different it is.
+ We classify a data set into N different classes.
    1. The entropy is 0 if all of the classes in the data are same
    2. The entropy is high if they are all different.

### Calculating Entropy

E(s) = -p1*ln(p1) -....-pn*ln(pn)

pi represent the proportion of data label for each class

### Demo with Python

I'll load data from titanic dataset. using pandas for conversion of csv file to DataFrame. 

```python
import numpy as np
import pandas as pd
from sklearn import tree

input_data = './titanic.csv';
df = pd.read_csv(input_data,sep='\\t',header=0)
```

```python
df.head()
```

Scikit-learn needs everything to be numerical for decision trees to work. So we will map categorical fields to numerical 

```python
binary = {'male':1,'female':0}
df['Sex']=df['Sex'].map(binary)
df.head()
```

```python
features = ['Sex','SibSp','Parch','Fare']
features
```

```python
y = df['Survived']
X = df[features]
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X,y)
```

```python
from IPython.display import Image  
from sklearn.externals.six import StringIO  
import pydotplus

dot_data = StringIO()  
tree.export_graphviz(clf, out_file=dot_data,  
                         feature_names=features)  
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())  
```
